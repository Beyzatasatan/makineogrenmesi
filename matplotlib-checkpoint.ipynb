```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y)
plt.title("Basit çizgi grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
kategori=["a","b","c","d"]
values=[3,7,5,12]
plt.bar(kategori,values ,color="#f542ef")
plt.title("Basit çubuk grafiği",color="#42f5ec",fontsize=50)
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
kategori=["a","b","c","d"]
values=[3,7,5,12]
plt.barh(kategori,values ,color="#f542ef")
plt.title("Basit çubuk grafiği",color="#42f5ec",fontsize=50)
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid()
plt.show()
```

```python
import numpy as np
kategori=["a","b","c","d"]
values1=[3,7,5,12]
values2=[4,6,8,10]
values3=[9,10,11,13]
x=np.arange(len(kategori))
print(x)
plt.bar(x-0.2,values1,width=0.2,label="set1")
plt.bar(x ,values2,width=0.2,label="set2")
plt.bar(x+0.2 ,values3,width=0.2,label="set3")
plt.xticks(x,kategori)
plt.title("ggggggggggg")
plt.xlabel("kategori")
plt.ylabel("degerler")
plt.legend()
plt.show()
```

```python
plt.bar(kategori,values1,label="set1")
plt.bar(kategori,values2,bottom=values1,label="set2")

plt.title("yığılmış çubuk grafiği")
plt.xlabel("kategori")
plt.ylabel("degerler")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
a=["elma","muz","kiraz","portakal"]
sizes=[15,30,45,10]
plt.pie(sizes,labels=a,autopct='%1.1f%%')
plt.title("meyve dağıtılımı")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=10,histtype="bar")
plt.title("histogram")
plt.ylabel("frekans")
plt.xlabel("degerler")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,25,30,35]
plt.scatter(x,y, marker="*",color="red",linewidth=5)
plt.title("dağılım")
plt.ylabel("y değeri")
plt.xlabel("x degeri")

plt.show()
```

```python
colors=[10,20,30,40,50]
sizes=[100,200,300,400,500]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap="viridis")
plt.colorbar()
plt.title("özelleştirilmiş dağılım")
plt.ylabel("y değeri")
plt.xlabel("x degeri")

plt.show()
```
