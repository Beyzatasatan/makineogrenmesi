```python
import pandas as pd
import numpy as np
sozluk={"İsim":pd.Series(["ada","cem","sibel","ahmet","mehmet","ali","veli","ayşe","hüseyin","necmi","nalan","namık"]),
        "meslek":pd.Series(["işçi","işçi","memur","serbest","serbest",None,None,"sigortacı","işsiz",None,None,"memur"]),
        "tarih":pd.Series(["05.06.2024","03.02.2023","03.02.2023","11.12.2020","11.12.2020",None,None,None,"03.02.2023",None,"11.12.2020","11.12.2020"]),
        "yas":pd.Series([21,24,25,44,31,27,35,33,42,29,41,43]),
        "cocuksayısı":pd.Series([None,None,None,None,None,1,2,0,None,None,None,None,]),
        "puan":pd.Series([89,87,77,55,70,79,73,79,54,92,61,69])}
df=pd.DataFrame(sozluk)
df
```

```python
def eksik_deger_tablosu(df):
    eksik_deger=df.isnull().sum()
    eksik_deger_yuzde=100*df.isnull().sum()/len(df)
    eksik_deger_tablo=pd.concat([eksik_deger, eksik_deger_yuzde],axis=1)
    son=eksik_deger_tablo.rename(columns={0:"eksik degerler",1:"% degeri"})
    return son
eksik_deger_tablosu(df)
```

```python
tr=len(df)*.3
df.dropna(thresh=tr,axis=1,inplace=True)
df#yüzdeliği 30 üzreninde olan sütünu sil
```

```python
df["meslek"]=df["meslek"].fillna("diğer")
print(df["tarih"].unique()[0])
df["tarih"]=df["tarih"].fillna(df["tarih"].unique()[0])
df
```

```python
#df["gecti"]=df["puan"]>70
def basari(puan):
    return(puan>=70)
df["geçti"]=df["puan"].apply(basari)
df
```

```python
df["yıl"]=pd.to_datetime(df["tarih"]).dt.year

df
```

```python
def durum(meslek):
    if meslek == "işçi":
        return 1
    elif  meslek == "memur":
        return 2
    elif  meslek == "serbest":
        return 3
    elif  meslek == "diğer":
        return 4
    elif  meslek == "sigortacı":
        return 5
    elif  meslek == "işsiz":
        return 6

df["meslek_kod"] = df["meslek"].apply(durum)
df
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df["geçti"]=label_encoder.fit_transform(df["geçti"])
df["meslek"]=label_encoder.fit_transform(df["meslek"])
df
```

```python
x=df[["puan"]].values.astype(float)
min_max_scaler=preprocessing.MinMaxScaler()
x_scaled=min_max_scaler.fit_transform(x)
df["puan2"]=pd.DataFrame(x_scaled)
df
```

```python
x=df[["puan"]].values.astype(float)
min_max_scaler=preprocessing.StandardScaler()
x_scaled=min_max_scaler.fit_transform(x)
df["puan3"]=pd.DataFrame(x_scaled)
df
```
